```python
import re
robustre = re.compile("robust(0.5|-2|2)")
robustre.search(fn)
```

```python
line_cubes = {}
for fn in glob.glob("*BN_only*medsub*.image.pbcor.fits"):
    cube = SpectralCube.read(fn)
    lines = cube.find_lines(chemical_name='Hydrogen Recombination')
    robust = robustre.search(fn).groups()[0]
    if len(lines)>0:
        print(f"{fn}: {lines['Freq-GHz(rest frame,redshifted)'][0]}")
        frq = u.Quantity(lines['Freq-GHz(rest frame,redshifted)'][0], u.GHz)
        subcube = cube.with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=frq).spectral_slab(-10*u.km/u.s,
                                                                                                               60*u.km/u.s)
        subcube.beam_threshold = 1
        mx = subcube.max(axis=0)
        mx.quicklook()
        pl.gca().set_title(f"{lines['Species'][0]}_{lines['Resolved QNs'][0]}")
        if mx.max() > stats.mad_std(mx) * 3:
            line_cubes[f"{lines['Species'][0]}_{lines['Resolved QNs'][0]}_robust{robust}"] = subcube
```

```python
from astropy import wcs
```

```python
y, x = np.mgrid[:128, :128]
x, y = h30alpha.wcs.celestial.wcs_pix2world(x.ravel(), y.ravel(), 0)
g_init.x_mean, g_init.y_mean = h30alpha.wcs.celestial.wcs_pix2world(70, 70, 0)
pixscale = wcs.utils.proj_plane_pixel_scales(h30alpha.wcs.celestial)
g_init.x_stddev, g_init.y_stddev = 3*pixscale
centroids_30 = [fit_g(g_init, x, y, slc.value.ravel()) for slc in h30alpha]
```

```python
y, x = np.mgrid[:128, :128]
x, y = h42alpha.wcs.celestial.wcs_pix2world(x.ravel(), y.ravel(), 0)
g_init.x_mean, g_init.y_mean = h42alpha.wcs.celestial.wcs_pix2world(70, 70, 0)
pixscale = wcs.utils.proj_plane_pixel_scales(h42alpha.wcs.celestial)
g_init.x_stddev, g_init.y_stddev = 3*pixscale
centroids_42 = [fit_g(g_init, x, y, slc.value.ravel()) for slc in h42alpha]
```

```python
y, x = np.mgrid[:128, :128]
centroids_33 = [fit_g(g_init, x, y, slc.value) for slc in h33beta]
```

```python
pl.imshow(centroids_30[10](x,y).reshape(128,128))
pl.imshow(h30alpha[10].value - centroids_30[10](x,y).reshape(128,128))
h30alpha[10].value.max()
```

```python
x30 = np.array([gf.x_mean.value-g_init.x_mean.value for gf in centroids_30])*3600
y30 = np.array([gf.y_mean.value-g_init.y_mean.value for gf in centroids_30])*3600
pl.scatter(x30,
           y30,
           c=h30alpha.spectral_axis.value, marker='o')
pl.scatter(np.array([gf.x_mean.value-g_init.x_mean.value for gf in centroids_42])*3600,
           np.array([gf.y_mean.value-g_init.y_mean.value for gf in centroids_42])*3600,
           c=h42alpha.spectral_axis.value, marker='s')
#pl.axis((np.min([gf.x_mean.value for gf in centroids_30]), np.max([gf.x_mean.value for gf in centroids_30]),
#        np.min([gf.y_mean.value for gf in centroids_30]), np.max([gf.y_mean.value for gf in centroids_30])))
pl.axis([0.00, 0.03, -0.033, 0.0])
pl.colorbar()
sep = ((np.median(x30[h30alpha.spectral_axis < 25*u.km/u.s]) - np.median(x30[h30alpha.spectral_axis > 25*u.km/u.s]))**2 +
       (np.median(y30[h30alpha.spectral_axis < 25*u.km/u.s]) - np.median(y30[h30alpha.spectral_axis > 25*u.km/u.s]))**2)**0.5

pl.xlabel("RA offset (arcsec)")
pl.ylabel("Dec offset (arcsec)")
print(f"Separation between low and high velocity: {sep} arcsec = {sep*400} AU")
```
